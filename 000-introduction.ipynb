# Machine Learning

(approaches to image processing)

- manual, heuristic construction
- principled, model-based construction
- machine learning

(machine learning)

You need:

- large number of examples of inputs plus desired outputs
- a learning algorithm
- lots of compute power

You get:

- an automatically constructed "program" that approximates the input/output relationship
- a characterization of the expected error rate

(other names)

Machine learning is central to modern computer science, so people have invented lots of names for it to make it seem more related to what they are working on:

- machine learning
- pattern recognition
- data mining
- "big data"
- analytics
- parametric and non-parametric statistics
- ...

All of these use a core set of common techniques, but terminology and applications differ.

(machine learning on images)

- image classification (image to class)
- image segmentation (image to regions)
- object recognition (parts-of-image to classes)
- image processing (image to image)

(kinds of machine learning algorithms)

Machine learning can be based on a number of approaches:

- statistical models of data generation (generative)
- mathematical models of decision boundaries (discriminative)
- mathematical models of neurons (neural networks)
- logical rules (expert systems, Bayesian networks)
- past examples and memory (case based reasoning)

Despite the different approaches, they often give similar performance.

(this course)

- nearest neighbor classifiers
- perceptrons
- linear least squares, logistic regression
- nonlinear classifiers
- generative classifiers (Gaussians), decision regions
- unsupervised learning (k-means, PCA)

```python
# Python example: quicksort
from random import choice,sample

def quicksort(l):
    if len(l)<2: return l
    pivot = choice(l)
    lo = [x for x in l if x<pivot]
    eq = [x for x in l if x==pivot]
    hi = [x for x in l if x>pivot]
    return quicksort(lo)+eq+quicksort(hi)

```
